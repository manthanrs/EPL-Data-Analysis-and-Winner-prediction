## CREATING OUR INITIAL MACHINE LEARNING MODEL

we need to figure out how we determine the accuracy of the model, this is actually really important choice, we are going to try couple of different matrics and which one makes more sense.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
predictors=['venue_code', 'opponent_code', 'day_code', 'hour']
```

```python
rf.fit(train_data[predictors], train_data['target'])
```

```python
y_hat = rf.predict(test_data[predictors])
```

```python
#accuracy_score--> what percentage of time was your predictions accurate
from sklearn.metrics import accuracy_score

acc = accuracy_score(test_data['target'], y_hat)
print(acc)
```

#Improving precision with rolling averages

split our matches data frame up by teams, cause what we want to do is compute rolling averages for teams performance like how did this team do in past few games

for example, if we are in match week 4, how did mancity do in the previous three match weeks and then we can pass that into the algorithm and the algorithm can use that information to make better predictions about how they will do in match week 4

```python
grouped_matches = matches.groupby('team')
```

```python
group = grouped_matches.get_group('Manchester United')
group
```

```python
def rolling_avgs(group, cols, new_cols):
  group = group.sort_values('date')
  rolling_stats = group[cols].rolling(3, closed='left').mean()
  group[new_cols] = rolling_stats
  group = group.dropna(subset=new_cols)
  return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
new_cols
```

```python
rolling_avgs(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby('team').apply(lambda x: rolling_avgs(x, cols, new_cols))
```

```python
matches_rolling
```

```python
#we have got a extra level of index on the left, we are going to drop it
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
#this is important because we need unique values for each index
matches_rolling.index = range(matches_rolling.shape[0])
```

  # RETRAINING OUR MACHINE LEARNING MODEL

  now we have some new predictors and we can make a new set of predictions using this predictors

  we got a precision hike from 46-47% to 59-60%

```python
def make_predictions(data, predictors):
  train_data = data[data['date'] < '2022-01-01']
  test_data = data[data['date'] > '2022-01-01']
  rf.fit(train_data[predictors], train_data['target'])
  y_hat = rf.predict(test_data[predictors])
  combined = pd.DataFrame(dict(actual=test_data['target'], predicted=y_hat), index=test_data.index)
  precision = precision_score(test_data['target'], y_hat)
  return combined, precision # Explicitly return the values
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
precision
```
